# Aufgabe 1
#### In der ersten Aufgabe sollen rote smarties auf Bildern gezählt werden. Dazu wird eine Funktion erstellt, die definierte Farbräume erkennt und erkannte Objektkonturen glättet. Anschließend werden die erkannten Objektkonturen für jede Farbe gezählt. Die Funktion wurde so gestaltet, dass diverse verschiedene Farbräume einfach hinzugefügt werden können und beim Funktionsaufruf ausgewählt werden können. Außerdem wird das untersuchte Bild mit den markierten erkannten Smarties/Objekten ausgegeben. So lässt sich leicht erkennen ob die Farben richtig erkannt wurden. 
#### Im folgenden Beispiel wurden rote und blaue smarties gezählt.

```python
# Import OpenCV
import cv2
# Since OpenCV is based on numpy:
import numpy as np
# Import pyplot (we will need this to plot with the notebook)
import matplotlib.pyplot as plt
# Import ColorThief
from colorthief import ColorThief
# Helper function to provide plots in true colors in the notebook
def imshow(title, im):
    plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
    plt.title(title)
```

```python
def smarties_zählen(bildpfad, ausgewählte_farben):
    bild = cv2.imread(bildpfad)

    if bild is None:
        print("Fehler: Das Bild konnte nicht geladen werden.")
        return

    # Konvertiere das Bild zu HSV (Hue, Saturation, Value)
    hsv_bild = cv2.cvtColor(bild, cv2.COLOR_BGR2HSV)

    # Definiere die Farbintervalle für verschiedene Smartie-Farben
    farbintervalle = {
        'rot': ([0, 100, 100], [10, 255, 255]),
        'blau': ([110, 50, 50], [130, 255, 255]),
        'gruen': ([40, 40, 40], [80, 255, 255]),
        # Hier kannst du weitere Farben hinzufügen
    }

    # Überprüfe, ob die ausgewählten Farben gültig sind
    ungültige_farben = [farbe for farbe in ausgewählte_farben if farbe not in farbintervalle]
    if ungültige_farben:
        print(f"Warnung: Die Farben {', '.join(ungültige_farben)} wurden nicht erkannt.")
        return {farbe: 0 for farbe in ausgewählte_farben}

    ergebnisse = {}

    for farbe in ausgewählte_farben:
        farbuntergrenze, farbobergrenze = farbintervalle[farbe]

        # Erzeuge die Farbmaske für die Smarties
        maske = cv2.inRange(hsv_bild, np.array(farbuntergrenze), np.array(farbobergrenze))

        # Erzeuge den Konturfilter
        konturen, _ = cv2.findContours(maske, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Glättung
        gültige_konturen = [cnt for cnt in konturen if cv2.contourArea(cnt) > 100]

        # Zeichne Konturen auf das Bild
        cv2.drawContours(bild, gültige_konturen, -1, (0, 255, 0), 2)

        ergebnisse[farbe] = len(gültige_konturen)

    bild_rgb = cv2.cvtColor(bild, cv2.COLOR_BGR2RGB)

    plt.imshow(bild_rgb)
    plt.title(bildpfad)
    plt.axis('off')
    plt.show()
    

    return ergebnisse

# Beispielaufruf für rote und blaue Smarties
#bildpfad = 'smarties0.png'
#ausgewählte_farben = ['rot', 'blau']
#ergebnisse = smarties_zählen(bildpfad, ausgewählte_farben)
#for farbe, anzahl in ergebnisse.items():
#    print(f'Anzahl {farbe} Smarties: {anzahl}')
```

```python
bildpfad = 'smarties0.png'
ausgewählte_farben = ['rot', 'blau']
ergebnisse = smarties_zählen(bildpfad, ausgewählte_farben)
for farbe, anzahl in ergebnisse.items():
    print(f'Anzahl {farbe} Smarties: {anzahl}')
```

#### Die Funktion hat 17 rote und 16 blaue Smarties erkannt. Die roten Smarties wurden sehr zuverlässig erkannt, ohne rosa Smarties zu den roten zu zählen oder rote zu übersehen. Bei den blauen Smarties besteht evtl. noch Optimierungspotenzial. Es wurden sowohl die vier großen blauen als auch die kleineren lilanen Smarties als blau erkannt. Außerdem wurden sämtliche hellblauen Smarties nicht erkannt. Hier besteht also je nachdem welche Farben zu blau gezählt werden sollen Justierbedarf der Farbräume.

# Aufgabe 2: Challenge 1
#### Bei der ersten Challenge sollen große und kleine Smarties gezählt werden. Die erstellt Funktion verwendet HoughCircles um Runde Objekte zu erkennen und als Kreise anzunähern. Mittels radius_threshold wird ein Grenzradius definiert, der große von kleinen Smarties unterscheidet. Das untersuchte Bild wird intern in Graustufen umgewandelt, untersucht und anschließend wieder als farbiges Bild mit markierten erkannten Smarties ausgegeben. Die Paramter von HoughCircles müssen angepasst werden um möglichst alle Smarties zu erkennen. Dies stellt sich als Schwierigkeit dar, da insbesondere helle Smarties nicht erkannt werden. Trotzdem konnte ein zufriedenstellendes Ergebnis erreicht werden.

```python
def zähle_große_kleine_smarties(image_path, min_radius, max_radius, radius_threshold):
    img = cv2.imread(image_path)

    if img is None:
        print("Fehler: Das Bild konnte nicht geladen werden.")
        return

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    circles = cv2.HoughCircles(
        blurred,
        cv2.HOUGH_GRADIENT,
        dp=0.5,
        minDist=40,
        param1=45,
        param2=17,
        minRadius=min_radius,
        maxRadius=max_radius
    )

    if circles is not None:
        circles = np.uint16(np.around(circles))

        small_smarties = 0
        big_smarties = 0

        for i in circles[0, :]:
            radius = i[2]

            if radius < radius_threshold:
                small_smarties += 1
                cv2.circle(img, (i[0], i[1]), radius, (0, 0, 255), 2)
            else:
                big_smarties += 1
                cv2.circle(img, (i[0], i[1]), radius, (0, 255, 0), 2)

        print(f"Anzahl kleiner Smarties: {small_smarties}")
        print(f"Anzahl großer Smarties: {big_smarties}")

        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title('Smarties')
        plt.axis('off')
        plt.show()

        return {'kleine_smarties': small_smarties, 'große_smarties': big_smarties}
    else:
        print("Keine Smarties gefunden.")
        return {'kleine_smarties': 0, 'große_smarties': 0}

# Beispielaufruf
#image_path = 'smarties0.png'  # Ersetze dies durch den tatsächlichen Pfad zu deinem Bild
#min_radius=10
#max_radius=100
#radius_threshold=50
#ergebnisse = zähle_große_kleine_smarties(image_path, min_radius=10, max_radius=100, radius_threshold=50)
#print(ergebnisse)
```

```python
image_path = 'smarties0.png'  # Ersetze dies durch den tatsächlichen Pfad zu deinem Bild
min_radius=20
max_radius=100
radius_threshold=50
zähle_große_kleine_smarties(image_path, min_radius, max_radius, radius_threshold)
```

#### Es wurden 63 kleine und 24 große Smarties erkannt. Nur 3 gelbe Smarties im unteren Teil des Bildes wurden nicht erkannt. Dies kommt daher, dass HoughCircles mit Graustufen arbeitet und sich die gelben Smarties dann kaum vom Hintergrund abheben.
#### Durch weitere Bearbeitung des Bildes könnte dieses Problem evtl. behoben werden.

# Aufgabe 3: Challenge 2
#### In der Challenge 2 sollen Bilder miteinander verglichen und entfernte Smarties gezält werden. Hier könnte man die Bilder subtrahieren und übrig gebliebene smarties zählen. Da jedoch nicht nur Smarties entfernt sondern teilweise auch leicht verschoben wurden, funktioniert die Erkennung nicht zuverlässig. Wenn sie funktioniert bietet diese Methode jedoch den vorteil das angezeigt werden kann an welchen Stellen Smarties entfernt wurden.
#### Da dies hier aber zunächst nicht nötig ist wird eine Funktion verwendet, welche wie zuvor mit HoughCircles Smarties zählt, ohne jedoch zwischen großen und kleinen zu unterscheiden. Anschließend werden aus den gezählten Smarties die Differenzen von Bild zu Bild berechnet. Das Problem der nicht erkannten gelben Smarties besteht jedoch weiterhin, 

```python
def count_smarties(image_path, min_radius=10, max_radius=100):

    img = cv2.imread(image_path)

    if img is None:
        print("Fehler: Das Bild konnte nicht geladen werden.")
        return
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    

    # Finde Kreise im Bild
    circles = cv2.HoughCircles(
        blurred,
        cv2.HOUGH_GRADIENT,
        dp=0.5,
        minDist=40,
        param1=45,
        param2=17,
        minRadius=min_radius,
        maxRadius=max_radius
    )

    if circles is not None:
        circles = np.uint16(np.around(circles))

        # Zeichne die erkannten Kreise auf das Bild
        for i in circles[0, :]:
            cv2.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 2)

        # Anzeige des Bildes mit erkannten Smarties
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(f'Smarties auf {image_path}')
        plt.axis('off')
        plt.show()

        return len(circles[0])
    else:
        print("Keine Smarties gefunden.")
        return 0

# Beispielaufrufe für aufeinanderfolgende Bilder von smarties0 bis smarties7
smarties_counts = []
for i in range(7):
    image_path = f'smarties{i}.png'
    smarties_count = count_smarties(image_path)
    smarties_counts.append(smarties_count)

# Berechne die Differenz zwischen aufeinanderfolgenden Bildern
differences = [smarties_counts[i+1] - smarties_counts[i] for i in range(len(smarties_counts)-1)]

# Zeige die Anzahl der Smarties auf jedem Bild und die Differenzen
for i, count in enumerate(smarties_counts):
    print(f'Anzahl Smarties auf Bild {i}: {count}')

for i, diff in enumerate(differences):
    print(f'Differenz zwischen Bild {i} und Bild {i+1}: {diff}')
```

#### Es wurden alle Samrties gezählt und die Differenzen gebildet. Man sieht das auch hier die Erkennung bei allen Smarties mit Ausnahme der gelben gut funktioniert hat. Insgesamt wurden im Verlauf der Bilder 34 Smarties entfernt.
